```python
import os
import pandas as pd
import numpy as np
import pypsa
import sys
```

```python
n = pypsa.Network()
n.import_from_netcdf('elec_s.nc')
```

```python
file_name =  os.path.basename(sys.argv[0])
directory = os.path.dirname(os.path.realpath(file_name))+'/NTC_Data_2013'

skips = ([False]*3+[True]*24)*365 #removes empty lines in dataframe
NTCs = pd.DataFrame(index=n.generators_t.p.index)

for file in os.listdir(directory):
    df = pd.read_excel('NTC_Data_2013/'+file,header=3,index_col=0).dropna(axis=1,how='all')[skips[2:]]
    for i in df.columns:
        NTCs[i] = np.array(df[i])
        
NTCs = NTCs.astype(float)
NTCs.drop(columns = NTCs.columns[NTCs.columns.str.contains('UA')],inplace=True)
NTCs.columns = ['AT > CH', 'AT > CZ', 'AT > HU', 'AT > IT1', 'AT > SI', 'BA > HR', 'BA > ME', 'BA > RS', 'BE > FR', 'BE > NL',
                'BG > GR', 'BG > MK', 'BG > RO', 'BG > RS', 'CH > AT', 'CH > DE/LU', 'CH > FR', 'CH > IT1', 'CZ > AT',
                'CZ > DE/LU', 'CZ > PL', 'CZ > SK', 'DE/LU > PL', 'DE/LU > CH', 'DE/LU > CZ', 'DE/LU > DK1',
                'DE/LU > DK2', 'DE/LU > FR', 'DE/LU > NL', 'DE/LU > SE4', 'DK1 > DE/LU', 'DK1 > DK2', 'DK1 > NO2',
                'DK1 > SE3', 'DK2 > DE/LU', 'DK2 > SE4', 'EE > FI', 'EE > LV', 'ES > FR', 'ES > PT', 'FI > EE', 'FI > NO4',
                'FI > SE1', 'FI > SE3', 'FR > BE', 'FR > CH', 'FR > DE/LU', 'FR > ES', 'FR > GB', 'FR > IT1', 'HR > BA',
                'HR > HU', 'HR > RS', 'HR > SI', 'HU > AT', 'HU > HR', 'HU > RO', 'HU > RS', 'HU > SK', 'IT1 > AT', 'IT1 > CH',
                'IT1 > FR', 'IT4 > GR', 'IT1 > SI', 'LV > EE', 'LV > LT', 'MK > BG', 'MK > GR', 'MK > RS', 'NL > BE', 'NL > DE/LU',
                'NL > GB', 'NL > NO2', 'NO1 > NO2', 'NO1 > NO3', 'NO1 > NO5', 'NO1 > SE3', 'NO2 > DK1', 'NO2 > NL', 'NO2 > NO5',
                'NO3 > NO4', 'NO3 > SE2', 'NO4 > FI', 'NO4 > SE1', 'NO4 > SE2', 'PL > CZ', 'PL > DE/LU', 'PL > SE4',
                'PL > SK', 'RS > AL', 'RS > BA', 'RS > BG', 'RS > HR', 'RS > HU', 'RS > ME', 'RS > MK', 'RS > RO', 'SE1 > FI',
                'SE1 > NO4', 'SE1 > SE2', 'SE2 > NO3', 'SE2 > NO4', 'SE2 > SE3', 'SE3 > DK1', 'SE3 > FI', 'SE3 > NO1',
                'SE3 > SE4', 'SE4 > DE/LU', 'SE4 > DK2', 'SE4 > PL', 'SI > AT', 'SI > HR', 'SI > IT1', 'SK > CZ', 'SK > HU',
                'SK > PL', 'GB > FR', 'GB > IE', 'GB > NL']
NTCs.drop(columns = NTCs.columns[NTCs.columns.str.contains('RS')],inplace=True)

#Split NO1 > NO3 into two separate columns (no negatives allowed)
NTCs['NO3 > NO1'] = NTCs['NO1 > NO3']*-1
NTCs[NTCs < 0] = 0

#Drop FI > NO4 column, too low voltage
NTCs.drop(columns=['FI > NO4','NO4 > FI'],inplace=True)
```

```python
NTCs.to_csv('NTCs.csv')
```
