```python
import numpy as np
import pandas as pd
import pypsa  
import os 
import sys  
import datetime
from pyomo.environ import Constraint

file_name =  os.path.basename(sys.argv[0])
directory = os.path.dirname(os.path.realpath(file_name))+'/Output'
n_file = 'Output/' + [s for s in os.listdir(directory) if 'EU_Network_created' in s][0]

n4 = pypsa.Network()
n4.import_from_netcdf(n_file)

n_storage = pd.read_csv('Output/n_storage.csv',index_col=0)
n_storage.index = pd.to_datetime(n_storage.index)

Indices = np.load('Output/indices.npy')
```

```python
#Get the list of storage units with significant standard deviation over the course of the year
Storage_units = n4.storage_units.loc[n_storage.std(axis=0)>100000].index

n4.storage_units['cyclic_state_of_charge'] = False
```

```python
def storage_rule(model,su):
    final = n_storage.loc[n4.snapshots[hour],su]
    start = n_storage.loc[n4.snapshots[hour-1],su]
    if (final-start > 0) and ('hydro' in su):
        inflow = n4.storage_units_t.inflow.loc[n4.snapshots[hour],su]
        if start + inflow > final:
            return model.state_of_charge[su,n4.snapshots[hour]] >= final
        else:
            return model.state_of_charge[su,n4.snapshots[hour]] >= start + inflow
    else:
        return model.state_of_charge[su,n4.snapshots[hour]] >= final

def extra_functionality(network,snapshots):
    model = network.model
    model.storage = Constraint(Storage_units,rule=storage_rule)
```

```python
# Run optimization for desired timesteps
for i in Indices: 

    print('Hour: ',i)
    print(datetime.datetime.now())
    hour = i
    timesteps = np.array([hour])
    
    n4.storage_units['state_of_charge_initial'] = n_storage.iloc[hour-1]
    n4.lopf(n4.snapshots[timesteps],solver_name='gurobi',extra_functionality=extra_functionality,solver_options={'ResultFile':'model_2'+str(i)+'.ilp','TimeLimit':3600})

print(datetime.datetime.now())

n4.export_to_netcdf('Output/n4_10hour.nc')
n4.generators_t.p.iloc[Indices].to_csv('Output/n4_gen_10h.csv')
n4.storage_units_t.p.iloc[Indices].to_csv('Output/n4_storage_10h.csv')
```
